## Part 2: What does $Ax=b$ look like?

- $x$ solving $Ax=b$
- minimizes $\frac{1}{2}(Ax-b)^TA^{-1}(Ax-b) = \frac{1}{2}x^TAx - x^Tb +b^Tb $
  - maximizes $\;c -\frac{1}{2}(y-b)^TA^{-1}(y-b)$ for $y=Ax$

    which looks like the negative log of some particular kind of multivariate normal distribution with covariance matrix $A$ and mean $b$



```python
# wlog
b = np.zeros((2,1))

sd = np.array([4,1])
corr = .8
A = np.diag(sd)@np.array([[1,corr],[corr,1]])@np.diag(sd)
A_inv = np.linalg.inv(A)
A
```

```python
def L2_x(x):
    "for px1 vector x"
    return ((A@x).T@(A@x))**0.5

def squared_L2_x(x):
    "for px1 vector x"
    return (A@x).T@(A@x)

def mahalanobis_x(x):
    "for p-by-1 vector x"
    return ((A@x).T@A_inv@(A@x))**0.5

def squared_mahalanobis_x(x):
    "for p-by-1 vector x"
    return (A@x).T@A_inv@(A@x)

def L2_X(X):
    "for n-by-p X stacking n 1-by-p vectors x"
    return ( (X@A-b.T) * (X@A-b.T) ).sum(axis=1)**0.5

def squared_L2_X(X):
    "for n-by-p X stacking n 1-by-p vectors x"
    return ( (X@A-b.T) * (X@A-b.T) ).sum(axis=1)

def mahalanobis_X(X):
    "for n-by-p X stacking n 1-by-p vectors x"
    return ( (X@A-b.T)@A_inv * (X@A-b.T) ).sum(axis=1)**0.5

def squared_mahalanobis_X(X, l1_lambda=0):
    "for n-by-p X stacking n 1-by-p vectors x"
    return ( (X@A-b.T)@A_inv * (X@A-b.T) ).sum(axis=1) + l1_lambda*np.abs(X).sum(axis=1)
```

```python
# Have some data
# and suppose it's super pristine with no multicollinearity

n,p = 1000,8
z = stats.norm().rvs((n,p))
```

```python
# Now suppose you're a neural network layer so
# right now you're super nice and look like this

h1 = 2
W1 = stats.norm().rvs((p,h1))
b0 = stats.norm().rvs((h1,1))

def relu(W):
  W[W<0]=0
  return W

H = relu(z@W1 + b0.T)
H
```

```python
# And let's suppose the job of H is to engineer features
# for some downstream task that will look something like min_ùõΩ ||HùõΩ-y||_2^2
# [and we can even suppose H is doing a pretty good job!]

# min_ùõΩ (HùõΩ-y)^T (HùõΩ-y) = min_ùõΩ (HùõΩ)^T(HùõΩ) - 2(HùõΩ)^Ty

# So you take the derivative and set to zero...
# and omg it's our favorite Ax=b problem ever: H^TH ùõΩ = H^Ty
```

```python
# Ooph... that's not good is it?
np.round(H.T@H/n,3)
# [Sample a few times...
#  you'll get a bad one
#  and use that so the demonstrations shows how to fix it]
```

```python
# No: no it's not...

# Go ahead an try again if you want...
# Maybe you just got an unlucky random sample ;)
A = H.T@H
A_inv = np.linalg.inv(A)

n0, lim = 50, 50
x1, x2 = np.meshgrid(np.linspace(-lim,lim,n0), np.linspace(-lim,lim,n0))
X = np.concatenate([x1.reshape(n0**2, 1), x2.reshape(n0**2, 1)], axis=1)

plt.figure()
plt.contourf(x1, x2, L2_X(X).reshape(n0,n0))
plt.colorbar()
plt.figure()
plt.contourf(x1, x2, mahalanobis_X(X).reshape(n0,n0))
plt.colorbar()
plt.figure()
plt.contourf(x1, x2, squared_L2_X(X).reshape(n0,n0))
plt.colorbar()
plt.figure()
plt.contourf(x1, x2, squared_mahalanobis_X(X).reshape(n0,n0))
plt.colorbar();
```

```python
# But no... it's not that you're unlucky...
A/n
```

```python
# It's this: it's not centered and scaled

print(H.mean(axis=0))
H_c = H - H.mean(axis=0)
np.round(H_c.T@H_c/n,3) # cov
```

```python
# So let's center it...
H_c = H - H.mean(axis=0)
np.round(H_c.T@H_c/n,3) # cov confirmed
```

```python
# And scale it...
H_cs = H_c/H.std(axis=0)
np.round(H_cs.T@H_cs/n,3) # cor not cov
```

```python
# Looking better...
A_c = H_c.T@H_c
A_c/n  # cov(H)

A = A_c
A_inv = np.linalg.inv(A)

n0, lim = 50, 50
x1, x2 = np.meshgrid(np.linspace(-lim,lim,n0), np.linspace(-lim,lim,n0))
X = np.concatenate([x1.reshape(n0**2, 1), x2.reshape(n0**2, 1)], axis=1)

plt.figure()
plt.contourf(x1, x2, L2_X(X).reshape(n0,n0))
plt.colorbar()
plt.figure()
plt.contourf(x1, x2, mahalanobis_X(X).reshape(n0,n0))
plt.colorbar()
plt.figure()
plt.contourf(x1, x2, squared_L2_X(X).reshape(n0,n0))
plt.colorbar()
plt.figure()
plt.contourf(x1, x2, squared_mahalanobis_X(X).reshape(n0,n0))
plt.colorbar();
```
